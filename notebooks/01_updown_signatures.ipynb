```python
import os
import numpy as np
import pandas as pd


DATASETS = [
    "ibhi_dseq2",
    "katie_ebseq",
    "severin_edger",
    "yutaka_genmat"
]


def read_dataset(ds_name):
    file_path = os.path.join("../data/processed/00_signatures", f"{ds_name}.csv")
    return pd.read_csv(file_path)
```

```python
df = read_dataset("ibhi_dseq2")
df = df.dropna(subset=["gene_symbol"])
up = df[df['logfc'] > 0]
dw = df[df['logfc'] < 0]

up["gene_symbol"].to_csv("../data/processed/01_updown/ibhi_dseq2_up.txt", index=False, header=False)
dw["gene_symbol"].to_csv("../data/processed/01_updown/ibhi_dseq2_dw.txt", index=False, header=False)
```

```python
df = read_dataset("katie_ebseq")
df = df.dropna(subset=["gene_symbol"])
df['logfc'] = np.log2(df['realfc'])
up = df[df['logfc'] > 0]
dw = df[df['logfc'] < 0]

up["gene_symbol"].to_csv("../data/processed/01_updown/katie_ebseq_up.txt", index=False, header=False)
dw["gene_symbol"].to_csv("../data/processed/01_updown/katie_ebseq_dw.txt", index=False, header=False)
```

```python
df = read_dataset("yutaka_genmat")
df = df.dropna(subset=["gene_symbol"])
df["gene_symbol"].to_csv("../data/processed/01_updown/yutaka_genmat.txt", index=False, header=False)
```

```python
df = read_dataset("severin_edger")
df = df.dropna(subset=["gene_symbol"])
up = df[df['logfc'] > 0]
dw = df[df['logfc'] < 0]
up["gene_symbol"].to_csv("../data/processed/01_updown/severin_edger_up.txt", index=False, header=False)
dw["gene_symbol"].to_csv("../data/processed/01_updown/severin_edger_dw.txt", index=False, header=False)
```
